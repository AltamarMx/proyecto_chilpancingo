```python
import pandas as pd
import matplotlib.pyplot as plt
from readeplus_sql import *
```

```python
tmp = read_eplusql("../idf/cb/simulacion.sql",read='data') # by default reads everything
variables = tmp.vars
cb =  tmp.get_data(variables)

cb = cb.rename(columns={'Environment:Site Outdoor Air Drybulb Temperature (C)':'To'})
cb = cb.rename(columns={'THERMAL ZONE: ESTANCIA:Zone Mean Air Temperature (C)':'Tie'})

cb.plot(subplots=True,figsize=(12,10))
```

```python
tmp = read_eplusql("../idf/eps_exterior/simulacion.sql",read='data') # by default reads everything
variables = tmp.vars
eps =  tmp.get_data(variables)

eps = eps.rename(columns={'Environment:Site Outdoor Air Drybulb Temperature (C)':'To'})
eps = eps.rename(columns={'THERMAL ZONE: ESTANCIA:Zone Mean Air Temperature (C)':'Tie'})
eps.plot(subplots=True,figsize=(12,10))
```
